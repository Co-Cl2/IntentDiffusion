```python
# HINT3
import csv
import json

# ÂÆö‰πâ‰∏Ä‰∏™ÈõÜÂêàÊù•Â≠òÂÇ®Â∑≤ÁªèËßÅËøáÁöÑtextÔºåÈÅøÂÖçÈáçÂ§ç
seen_texts = set()
unique_labels = set()

# ÂÆö‰πâ‰∏Ä‰∏™ÂàóË°®Êù•Â≠òÂÇ®ËøáÊª§ÂêéÁöÑÊï∞ÊçÆ
filtered_data = []

# ÂÅáËÆæ‰Ω†ÁöÑCSVÊñá‰ª∂ÂêçÂ≠òÂÇ®Âú®‰∏Ä‰∏™ÂàóË°®‰∏≠
csv_files = ['/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/HINT3/curekart_test.csv', \
             '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/HINT3/curekart_train.csv', \
                '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/HINT3/powerplay11_test.csv', \
                    '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/HINT3/powerplay11_train.csv', \
                        '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/HINT3/sofmattress_test.csv', \
                            '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/HINT3/sofmattress_train.csv']

# ÈÅçÂéÜÊâÄÊúâCSVÊñá‰ª∂
for file_name in csv_files:
    with open(file_name, mode='r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            text = row['sentence']
            label = row['label']
            
            # Ê£ÄÊü•Ê†áÁ≠æÊòØÂê¶‰∏∫NO_NODES_DETECTEDÔºå‰ª•ÂèätextÊòØÂê¶ÈáçÂ§ç
            if label != 'NO_NODES_DETECTED' and text not in seen_texts:
                filtered_data.append({'text': text, 'label': label})
                seen_texts.add(text)
                unique_labels.add(label)

unique_labels=list(unique_labels)
# ËΩ¨Êç¢‰∏∫JSONÊ†ºÂºè
json_data = json.dumps(filtered_data, ensure_ascii=False, indent=4)

# ËæìÂá∫ÁªìÊûú
print(json_data)
print(f"‰∏ÄÂÖ±Êúâ {len(seen_texts)} ‰∏™Âè•Â≠ê„ÄÇ{len(unique_labels)}‰∏™Ê†áÁ≠æ")

output_json_file = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/HINT3/total.json'
with open(output_json_file, 'w', encoding='utf-8') as f_out:
    f_out.write(json_data)
```

```python
# SNIPS
import json

# ÂÆö‰πâÊñá‰ª∂Âêç
file1_text = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/SNIPS/test/seq.in'
file1_label = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/SNIPS/test/label'
file2_text = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/SNIPS/valid/seq.in'
file2_label = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/SNIPS/valid/label'
file3_text = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/SNIPS/train/seq.in'
file3_label = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/SNIPS/train/label'

files = [(file1_text,file1_label),(file2_text,file2_label),(file3_text,file3_label)]

output_json_name = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/SNIPS/total.json'

# ÂàùÂßãÂåñ‰∏§‰∏™ÂàóË°®ÔºåÂàÜÂà´Â≠òÂÇ®textÂíålabel
texts = []
labels = []
for file1,file2 in files:
    # ËØªÂèñÁ¨¨‰∏Ä‰∏™Êñá‰ª∂ÔºåËé∑Âèñtext
    with open(file1, 'r') as file1:
        for line in file1:
            # ÁßªÈô§Ë°åÂ∞æÁöÑÊç¢Ë°åÁ¨¶Âπ∂Ê∑ªÂä†Âà∞textsÂàóË°®
            texts.append(line.strip())

    # ËØªÂèñÁ¨¨‰∫å‰∏™Êñá‰ª∂ÔºåËé∑Âèñlabel
    with open(file2, 'r') as file2:
        for line in file2:
            # ÁßªÈô§Ë°åÂ∞æÁöÑÊç¢Ë°åÁ¨¶Âπ∂Ê∑ªÂä†Âà∞labelsÂàóË°®
            labels.append(line.strip())

# ÂêàÂπ∂textÂíålabel‰∏∫‰∏Ä‰∏™ÂàóË°®ÔºåÊØè‰∏™ÂÖÉÁ¥†ÊòØ‰∏Ä‰∏™Â≠óÂÖ∏
data = [{'text': text, 'label': label} for text, label in zip(texts, labels)]

# Â∞ÜÂêàÂπ∂ÂêéÁöÑÊï∞ÊçÆÂÜôÂÖ•JSONÊñá‰ª∂
with open(output_json_name, 'w') as output_file:
    json.dump(data, output_file, indent=4)

print(f'JSON file {output_json_name} has been created.')
```

```python
# PolicyIE
import json
import re

# ÂàùÂßãÂåñ‰∏Ä‰∏™Á©∫ÂàóË°®Êù•Â≠òÂÇ®ÂêàÂπ∂ÂêéÁöÑÊï∞ÊçÆ
merged_data = []

# ÂÆö‰πâ‰∏âÂØπÊñá‰ª∂ÁöÑÂêçÁß∞
file_pairs = [
    ('/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/PolicyIE/test.source', '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/PolicyIE/test.target'),
    ('/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/PolicyIE/train.source', '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/PolicyIE/train.target'),
    ('/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/PolicyIE/valid.source', '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/PolicyIE/valid.target')
]

# Ê≠£ÂàôË°®ËææÂºèÔºåÁî®‰∫éÂåπÈÖç[IN:xxx]Ê†ºÂºèÁöÑlabelÔºåÂπ∂Âè™Âèñxxx‰∏≠ÁöÑÁ¨¨‰∏Ä‰∏™ÂçïËØç
label_pattern = re.compile(r'\[IN:(\S+)')

# ÈÅçÂéÜÊñá‰ª∂ÂØπ
for text_file, label_file in file_pairs:
    # ËØªÂèñtextÊñá‰ª∂
    with open(text_file, 'r') as file:
        texts = file.readlines()
    
    # ËØªÂèñlabelÊñá‰ª∂
    with open(label_file, 'r') as file:
        labels = file.readlines()
    
    # Á°Æ‰øù‰∏§‰∏™Êñá‰ª∂ÁöÑË°åÊï∞Áõ∏Âêå
    assert len(texts) == len(labels), "The number of lines in text and label files must be the same."
    
    # ÈÅçÂéÜtextsÂíålabelsÔºåÊèêÂèñÂÜÖÂÆπÂπ∂Ê∑ªÂä†Âà∞merged_dataÂàóË°®
    for text, label in zip(texts, labels):
        # ÁßªÈô§Ë°åÂ∞æÁöÑÊç¢Ë°åÁ¨¶
        text = text.strip()
        label = label.strip()
        
        # ‰ΩøÁî®Ê≠£ÂàôË°®ËææÂºèÊèêÂèñlabel
        match = label_pattern.search(label)
        if match:
            label = match.group(1)  # ÊèêÂèñÊã¨Âè∑ÂÜÖÁöÑÂÜÖÂÆπ‰Ωú‰∏∫label
        else:
            # Â¶ÇÊûúÊ≤°ÊúâÂåπÈÖçÂà∞[IN:xxx]Ê†ºÂºèÁöÑlabelÔºåÂàôË∑≥ËøáËøô‰∏ÄË°å
            continue
        
        # Â¶ÇÊûúlabelÊòØ"Other"ÔºåÂàôËàçÂºÉÂØπÂ∫îÁöÑÊù°ÁõÆ
        if label != "Other]":
            merged_data.append({'text': text, 'label': label})

# Â∞ÜÂêàÂπ∂ÂêéÁöÑÊï∞ÊçÆÂÜôÂÖ•Êñ∞ÁöÑJSONÊñá‰ª∂
output_json_name = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/PolicyIE/total.json'
with open(output_json_name, 'w') as output_file:
    json.dump(merged_data, output_file, indent=4)

print(f'Merged JSON file {output_json_name} has been created.')
```

```python
# StackExchange
import json

# ÂàùÂßãÂåñ‰∏Ä‰∏™Á©∫ÂàóË°®Êù•Â≠òÂÇ®ÂêàÂπ∂ÂêéÁöÑÊï∞ÊçÆ
merged_data = []

# ÂÆö‰πâË¶ÅÂ§ÑÁêÜÁöÑÊñá‰ª∂Âêç
file_names = ['/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/StackExchange/AskUbuntuCorpus.json',\
                '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/StackExchange/WebApplicationsCorpus.json']

# ÈÅçÂéÜÊñá‰ª∂ÂàóË°®ÔºåËØªÂèñÊØè‰∏™Êñá‰ª∂
for file_name in file_names:
    with open(file_name, 'r') as file:
        # Âä†ËΩΩJSONÊï∞ÊçÆ
        data = json.load(file)
        
        # ÊèêÂèñsentencesÂàóË°®
        sentences = data.get('sentences', [])
        
        # ÈÅçÂéÜsentencesÂàóË°®ÔºåÊèêÂèñtextÂíåintent
        for sentence in sentences:
            text = sentence.get('text', '')
            intent = sentence.get('intent', '')
            
            # Â∞ÜÊèêÂèñÁöÑÊï∞ÊçÆÊ∑ªÂä†Âà∞merged_dataÂàóË°®
            merged_data.append({'text': text, 'label': intent})

# Â∞ÜÂêàÂπ∂ÂêéÁöÑÊï∞ÊçÆÂÜôÂÖ•Êñ∞ÁöÑJSONÊñá‰ª∂
output_json_name = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/StackExchange/total.json'
with open(output_json_name, 'w') as output_file:
    json.dump(merged_data, output_file, indent=4)

print(f'Merged JSON file {output_json_name} has been created.')
```

```python
import json
import random

# ÂÅáËÆæÊñá‰ª∂Âêç‰∏∫data.jsonÔºåÂåÖÂê´ÊÇ®Êèê‰æõÁöÑJSONÊï∞ÊçÆ
input_path = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/Bank77/total.json'
selected_output_path = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/Bank77/5-shot/train.json'
remaining_output_path = '/data/mashuotian/IntentDiffusion/Diffusion-LM/datasets/Intent-Classifier/Bank77/5-shot/test.json'

# ËØªÂèñJSONÊñá‰ª∂
with open(input_path, 'r') as file:
    data = json.load(file)

# ÊåâlabelÂàÜÁªÑÊï∞ÊçÆ
label_groups = {}
for item in data:
    label = item["label"]
    if label not in label_groups:
        label_groups[label] = []
    label_groups[label].append(item)

# Á°Æ‰øùÊúâË∂≥Â§üÁöÑÁ±ªÂà´
if len(label_groups) < 27:
    print("Not enough labels available.")
else:
    # ÈöèÊú∫ÈÄâÊã©27‰∏™Á±ªÂà´
    selected_labels = random.sample(list(label_groups.keys()), 27)

    # ÂàùÂßãÂåñËÆ≠ÁªÉÈõÜÂíåÊµãËØïÈõÜ
    selected_data = []
    remaining_data = []

    for label in selected_labels:
        items = label_groups[label]
        if len(items) < 5:
            print(f"Label {label} has less than 5 items, count: {len(items)}")
            selected_data.extend(items)  # Â¶ÇÊûú‰∏çË∂≥5‰∏™ÔºåÂÖ®ÈÉ®Âä†ÂÖ•ËÆ≠ÁªÉÈõÜ
        else:
            selected_data.extend(random.sample(items, 5))  # ÈöèÊú∫Âèñ5Êù°Êï∞ÊçÆ‰Ωú‰∏∫ËÆ≠ÁªÉÈõÜ
            remaining_data.extend([item for item in items if item not in selected_data])  # ÂèñÂâ©‰∏ãÁöÑÊï∞ÊçÆ‰Ωú‰∏∫ÊµãËØïÈõÜ

    # Êâì‰π±ÈÄâÂèñÁöÑËÆ≠ÁªÉÊï∞ÊçÆÈ°∫Â∫è
    random.shuffle(selected_data)
    random.shuffle(remaining_data)

    # Â∞ÜÈÄâÂèñÁöÑËÆ≠ÁªÉÊï∞ÊçÆÂÜôÂÖ•Êñ∞ÁöÑJSONÊñá‰ª∂
    with open(selected_output_path, 'w') as file:
        json.dump(selected_data, file, indent=4)

    # Â∞ÜÊú™Ë¢´ÈÄâÂèñÁöÑÊï∞ÊçÆÂÜôÂÖ•Âè¶‰∏Ä‰∏™JSONÊñá‰ª∂
    with open(remaining_output_path, 'w') as file:
        json.dump(remaining_data, file, indent=4)

    print(f"Selected and shuffled data has been written to {selected_output_path}")
    print(f"Remaining data has been written to {remaining_output_path}")
```
